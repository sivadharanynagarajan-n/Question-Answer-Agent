#### Installing the neccessary Library

```python
import os
from google.adk.agents import Agent
from google.adk.models.google_llm import Gemini
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types
import google.generativeai as genai
```

#### Getting the api key for the model

```python
genai.configure(api_key= os.getenv("GEMINI_API_KEY"))
```

#### Configure the retry attempts to manage unexpected hurdles

```python
retry_con = types.HttpRetryOptions(
    attempts = 5,
    exp_base = 7,
    initial_delay = 1,
    http_status_codes = [429, 500, 503, 504]
)
```

#### Defining the agent with all the necessary components(parameters)

```python
first_agent = Agent(
    name = "my_first_assistant",
    model = Gemini(
        model = "gemini-2.5-flash-lite",
        retry_options = retry_con
    ),
    description = "Building a first agent that answers general question",
    instruction = "You are a agent that answers question. You use google search in case you are unsure and return the final output",
    tools = [google_search]
)
print("Agent Defined Successfully")
```

#### Defining or initiating a runner that is a orchestrator

```python
runner = InMemoryRunner(agent = first_agent)

print("Runner created successfully")
```

#### Getting input from user and running it in the agent to give the answer 

```python
response = await runner.run_debug("How is the weather in chennai recently")
```
